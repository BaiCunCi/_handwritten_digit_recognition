```python

#导入训练集
from sklearn import datasets

#导入算法SVM
from sklearn import svm

#导入可视化工具
import matplotlib.pyplot as plt

#导入可以加载图片并处理
from PIL import Image

#导入处理矩阵向量
import numpy as np
```

```python
#题目来源
digits=datasets.load_digits()

# print(digits)

#查看图片数据（旧问题）
images_data=digits['images']
#print(images_data)
print("问题：",len(images_data))
#查看答案
images_title=digits['target']
print("答案：",len(images_title))
```

```python
plt.figure(figsize=(30,15))
for i in range(30):
    #设置子图(3行10列)
    plt.subplot(3,10,i+1)
    plt.imshow(images_data[i],cmap=plt.cm.gray_r)
#print(images_data[5])
#print(" 答案：",images_title[5])
#查看答案
    plt.title('%s'%(images_title[i]))
```

```python
#使用SVM算法生成方法
#问题X
X=digits['data']
#print(digits['data'].shape)
#答案Y
Y =digits['target']
```

```python
classifier=svm.SVC(gamma=0.001)
#学习
classifier.fit(X,Y)
```

```python
#模拟考
newX=digits['images'][7].reshape(1,-1)
#print(newX)
print("答案：",digits['target'][7])
answer=classifier.predict(newX)
print("机器学习：",answer)
```

```python
#高考 4
#导入图片
img=Image.open("./data/4.png").convert("L")
#创建8*8全0矩阵，int

img_data =np.zeros((8,8),dtype=np.int)
#print(img_data)
#转化为矩阵
for i in range(8):
    for j in range(8):
        img_data[i][j]=np.abs(16-img.getpixel((j,i))*16/255)
print(img_data)
plt.imshow(img_data,cmap=plt.cm.gray_r)
```

```python
ans=classifier.predict(img_data.reshape(1,-1))
print("机器学习结果：",ans)
```
